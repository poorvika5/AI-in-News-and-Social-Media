1. **Advancements in Language Representation Models**

This advancement has enabled more nuanced sentiment analysis and topic extraction from unlabelled text, making it a powerful tool in analyzing news headlines and social media posts.
2. **Sentiment Analysis in Financial Markets**

4. **News Framing and Public Opinion**
The role of media in shaping public opinion through news framing has been well-documented. Entman (1993) defines framing as the selective emphasis of certain aspects of a perceived reality to promote specific interpretations. Studies have shown that media coverage influences public attitudes towards new technologies, including AI (Anderson et al., 2005; Nelkin, 1987). Understanding media framing is crucial for assessing how AI-related topics are perceived and discussed in the public sphere.
5. **Crowdsourcing and Quantitative Analysis**
Crowdsourcing has emerged as a valuable tool for enhancing quantitative analysis. Studies have utilized crowdsourced annotations to identify themes and sentiments in large text corpora (Fast et al., 2016; Danescu-Niculescu-Mizil et al., 2013). This approach allows for more nuanced and comprehensive sentiment analysis, which can be applied to understanding public opinion on AI and other emerging technologies.
6. **Aspect-Based Sentiment Analysis**
Aspect-based sentiment analysis (ABSA) has been advanced by recent developments in language models like BERT. Sun et al. (2019) and Li et al. (2019) demonstrated the effectiveness of fine-tuning BERT for ABSA tasks, enhancing the ability to detect sentiment related to specific aspects of a text. This is particularly relevant for analyzing detailed sentiment in news headlines and social media posts.
7. **Zero-Shot and Cross-Lingual Learning**
In cases where labeled data is scarce, zero-shot and cross-lingual learning methods offer promising alternatives. Jebbara and Cimiano (2019) explored zero-shot cross-lingual approaches for sentiment analysis, addressing the challenges of limited annotated data in specific languages. This technique can be applied to extend sentiment analysis models to a broader range of languages and contexts.

Text Cleaning:

Removed URLs and email addresses.
Removed special characters and numbers.
Converted text to lowercase.
Removed punctuation and numbers.
Tokenized the text and removed stop words. Lemmatized the tokens to their base forms.
Included additional common words like ‘say’, ‘could’, ‘would’, ‘one’, ‘also’, and ‘many’ for removal.

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
additional_stopwords = set(['say', 'could', 'would', 'one', 'also', 'many'])

def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S*@\S*\s?', '', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = ''.join([char for char in text if not char.isdigit()])
    tokens = word_tokenize(text)
    filtered_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and word not in additional_stopwords and len(word) > 2 and word.isascii()]
    return filtered_tokens

df_cleaned['FilteredTokens'] = df_cleaned['Headline'].apply(clean_text)
df_cleaned['TokenCount'] = df_cleaned['FilteredTokens'].apply(len)
```

Structural Errors and Outliers:

Calculated the number of sentences in headlines and leads.
Performed Part-of-Speech (POS) tagging and counted nouns.
Computed descriptive statistics for token count, sentence count, and noun count.
Identified and removed rows with unrealistic token or sentence counts.
Used the Interquartile Range (IQR) method to detect and remove outliers.
Standardized and normalized the TokenCount and NounCount columns.

```python
# Correct calculation of SentenceCount
df_cleaned['SentenceCount_Headline'] = df_cleaned['Headline'].apply(lambda x: len(sent_tokenize(x)))
df_cleaned['SentenceCount_Lead'] = df_cleaned['Lead '].apply(lambda x: len(sent_tokenize(x)))
df_cleaned['SentenceCount'] = df_cleaned['SentenceCount_Headline'] + df_cleaned['SentenceCount_Lead']
```

```python
# Perform POS tagging and count nouns
df_cleaned['POS'] = df_cleaned['FilteredTokens'].apply(pos_tag)
df_cleaned['NounCount'] = df_cleaned['POS'].apply(lambda x: sum(1 for word, pos in x if pos.startswith('NN')))
```

```python
# Compute descriptive statistics
desc_stats = df_cleaned[['TokenCount', 'SentenceCount', 'NounCount']].describe()
print("Descriptive Statistics:\n", desc_stats)
```

```python
# Check for structural errors (e.g., check if token counts or sentence counts are unrealistic)
structural_errors = df_cleaned[(df_cleaned['TokenCount'] > 50) | (df_cleaned['SentenceCount'] > 5)]
print("Structural Errors:\n", structural_errors)
```

```python
# Removing structural errors
df_cleaned = df_cleaned[(df_cleaned['TokenCount'] <= 50) & (df_cleaned['SentenceCount'] <= 5)]
```

```python
# Outlier detection using IQR
Q1 = df_cleaned[['TokenCount', 'SentenceCount', 'NounCount']].quantile(0.25)
Q3 = df_cleaned[['TokenCount', 'SentenceCount', 'NounCount']].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define outliers as points outside 1.5 * IQR
outliers = ((df_cleaned[['TokenCount', 'NounCount']] < (Q1[['TokenCount', 'NounCount']] - 1.5 * IQR[['TokenCount', 'NounCount']])) | (df_cleaned[['TokenCount', 'NounCount']] > (Q3[['TokenCount', 'NounCount']] + 1.5 * IQR[['TokenCount', 'NounCount']]))).any(axis=1)
print(f"Number of outliers detected: {outliers.sum()}")
```

```python
# Removing outliers
df_cleaned = df_cleaned[~outliers]
```

```python
# Standardize the data (TokenCount and NounCount)
scaler = StandardScaler()
df_cleaned[['TokenCount', 'NounCount']] = scaler.fit_transform(df_cleaned[['TokenCount', 'NounCount']])
```

```python
# Normalize the data (TokenCount and NounCount)
normalizer = MinMaxScaler()
df_cleaned[['TokenCount', 'NounCount']] = normalizer.fit_transform(df_cleaned[['TokenCount', 'NounCount']])

# Updated descriptive statistics after removing structural errors, outliers, and standardization/normalization
updated_desc_stats = df_cleaned[['TokenCount', 'SentenceCount', 'NounCount']].describe()
print("Updated Descriptive Statistics:\n", updated_desc_stats)
```

```python
 # N-gram Analysis
import pandas as pd
from nltk import word_tokenize, ngrams
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
def get_ngrams(tokens, n=2):
    n_grams = ngrams(tokens, n)
    return [' '.join(grams) for grams in n_grams]


```

```python
def plot_ngrams(ngrams, title, n=10):
    # Count the frequency of each n-gram
    ngram_freq = Counter(ngrams)

    # Get the most common n-grams
    most_common = ngram_freq.most_common(n)
    ngram_labels, counts = zip(*most_common)

    # Create a DataFrame for plotting
    df = pd.DataFrame({'ngram': ngram_labels, 'count': counts})

    # Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='count', y='ngram', data=df, palette='viridis')
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('N-gram')
    plt.show()

# Extract tokens from cleaned DataFrame
bigrams_headline = [ngram for sublist in df_cleaned['FilteredTokens'] for ngram in get_ngrams(sublist, n=2)]
bigrams_lead = [ngram for sublist in df_cleaned['FilteredTokens'] for ngram in get_ngrams(sublist, n=2)]
trigrams_headline = [ngram for sublist in df_cleaned['FilteredTokens'] for ngram in get_ngrams(sublist, n=3)]
trigrams_lead = [ngram for sublist in df_cleaned['FilteredTokens'] for ngram in get_ngrams(sublist, n=3)]

# Plot the n-grams
plot_ngrams(bigrams_headline, 'Top Bigrams in Headlines')
plot_ngrams(bigrams_lead, 'Top Bigrams in Leads')
plot_ngrams(trigrams_headline, 'Top Trigrams in Headlines')
plot_ngrams(trigrams_lead, 'Top Trigrams in Leads')
```

Adding class weights to Logistic Regression and SVM models is a technique used to handle class imbalance in classification problems. This approach adjusts the cost function to place more emphasis on minority classes during training, thereby helping the model to learn from these underrepresented classes.

Why Add Class Weights
Handling Class Imbalance:

Class imbalance occurs when certain classes have significantly more samples than others. This can cause the model to be biased towards the majority class, leading to poor performance on minority classes.
Adding class weights compensates for this imbalance by giving more importance to the minority classes during the training process.
Improving Model Performance:

By incorporating class weights, the model becomes more sensitive to the minority classes, improving metrics like precision, recall, and F1-score for these classes.
This leads to a more balanced and fair performance across all classes.
Pros and Cons of Adding Class Weights
Pros
Improved Recall for Minority Classes:

The model is better able to detect minority class instances, increasing the recall for these classes.
Balanced Performance:

The model achieves a more balanced performance across all classes, not just the majority class, leading to better overall metrics like macro-averaged precision, recall, and F1-score.
Fairness:

Ensures that the minority classes are fairly represented, which is crucial in applications where all classes are equally important.
Cons
Potential Overfitting:

If the dataset is extremely imbalanced, the model might overfit to the minority class, leading to poor generalization on unseen data.
Increased Complexity:

Adjusting class weights adds complexity to the model training process and may require additional tuning to find the optimal weights.

```python
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def calculate_sentiment(headline):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_score = analyzer.polarity_scores(headline)['compound']
    return sentiment_score

# Apply sentiment analysis to each headline
df_cleaned['Sentiment'] = df_cleaned['Headline'].apply(calculate_sentiment)

# Comparative Analysis: CNN vs BBC
def comparative_analysis(df):
    # Token, sentence, and noun counts by channel
    channel_stats = df.groupby('ChannelName')[['TokenCount', 'SentenceCount', 'NounCount']].mean()

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    sns.barplot(x=channel_stats.index, y=channel_stats['TokenCount'])
    plt.title('Average Token Count by Channel')

    plt.subplot(1, 3, 2)
    sns.barplot(x=channel_stats.index, y=channel_stats['SentenceCount'])
    plt.title('Average Sentence Count by Channel')

    plt.subplot(1, 3, 3)
    sns.barplot(x=channel_stats.index, y=channel_stats['NounCount'])
    plt.title('Average Noun Count by Channel')

    plt.show()

    # Sentiment by channel
    channel_sentiment = df.groupby('ChannelName')['Sentiment'].mean()

    plt.figure(figsize=(8, 6))
    sns.barplot(x=channel_sentiment.index, y=channel_sentiment.values)
    plt.title('Average Sentiment by Channel')
    plt.xlabel('Channel')
    plt.ylabel('Average Sentiment')
    plt.show()

    # Most frequent words by channel
    cnn_words = [word for tokens in df[df['ChannelName'] == 'CNN']['FilteredTokens_Headline'] for word in tokens]
    bbc_words = [word for tokens in df[df['ChannelName'] == 'BBC']['FilteredTokens_Headline'] for word in tokens]

    cnn_word_freq = nltk.FreqDist(cnn_words)
    bbc_word_freq = nltk.FreqDist(bbc_words)

    common_cnn_words = cnn_word_freq.most_common(20)
    common_bbc_words = bbc_word_freq.most_common(20)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    sns.barplot(x=[word for word, freq in common_cnn_words], y=[freq for word, freq in common_cnn_words])
    plt.title('Most Common Words in CNN Headlines')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)

    plt.subplot(1, 2, 2)
    sns.barplot(x=[word for word, freq in common_bbc_words], y=[freq for word, freq in common_bbc_words])
    plt.title('Most Common Words in BBC Headlines')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)

    plt.show()

# Call the comparative analysis function
comparative_analysis(df_cleaned)
```
